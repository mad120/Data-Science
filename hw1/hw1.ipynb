```python
df = pd.read_csv("Data_Collection.tsv", sep="\t")
df
```

```python
df["trump_mention"] = df["content"].str.contains("(^|[^A-Za-z0-9])Trump([^A-Za-z0-9]|$)", regex=True, case=True)
len(df[df["trump_mention"] == True])
```

```python
finaldf=df[["tweet_id", "publish_date", "content", "trump_mention"]]
len(finaldf)
```

```python
finaldf.to_csv("dataset.tsv",sep="\t", index=False) #Output of Data annonation phase
```

```python
trumpTrueCount=len(df[df['trump_mention']==True])
trumpFalseCount=len(df[df['trump_mention']==False])
result= format(trumpTrueCount/(trumpFalseCount+trumpTrueCount), '.3f')
print(result)
```

```python
data = [['frac-trump-mentions', result] ] 
resdf = pd.DataFrame(data, columns = ['result', 'value']) 
resdf
```

```python
resdf.to_csv("results.tsv",sep="\t", index=False)  #Output of Analysis phase
```
